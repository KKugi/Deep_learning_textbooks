## 딥러닝을 가능케한 autograd

```python
x=torch.tensor([1.], requires_grad=True) # float 이여야해서 1.
print(x)
```

```python
x=torch.tensor([1.])
print(x)
print(x.requires_grad)
print()

x.requires_grad=True
print(x)
print(x.requires_grad)
```

```python
x=torch.tensor([1.], requires_grad=True)
y=x**2
print(y) # PowBackward0 가 붙음

print(x.grad)
y.backward() # requires_grad가 True인 변수에 대해서 미분, requires_grad=False 이면 에러 발생
print(x.grad) # y=x**2 를 미분한 2x의 x값에 1을 대입한 gradient 값
```

image.png

```python
x=torch.tensor([1.], requires_grad=True)
y=x**2
print(y)
y.retain_grad() # 이걸 하면 y.grad도 볼 수 있다.

z=3*y
print(z) # MulBackward0 가 붙어있다

"""
z는 y에다가 3을 곱했다: 실제로는 제곱을 했다, 곱했다 이 두개의 사실이 다 들어가 있음
y에 대한 미분이 아니라 x에 대한 미분
그러므로 x제곱하고 나서 3을 곱한 3x^2에 대해서 x로 미분한 결과인 6x에 1을 대입한 결과인 6이 x.grad에 저장되어 있음
"""

z.backward()
print(x.grad) # chain rule로 알아냄.
# print(y.grad) # warning! 중간건 안된다
# leaf Tensor은 requires_grad=True 인 녀석임
print(y.grad) # y.retain_grad() # 이걸 하면 y.grad도 볼 수 있다.
```

```python
x=torch.tensor([1.], requires_grad=True)
y=x**2
z=3*y

y.backward() # 이렇게 하면 y에서부터 뒤로 넘어감 (backward!) : x제곱을 미분한 2x에다가 1을 대입한 2가 나옴
print(x.grad)
```

```python
# detach 사용 용도
x=torch.tensor([1.], requires_grad=True)
y=x**2
z=y.detach() # x로 만든 것을 상수로 사용하고 싶은 것. y.requires_grad=False 뭐 이런 식으로 중간에 상수로 바꿀 수는 없다
w=y+z # x**2+1과 같다, x^2+x^2이 아님 => y.detach()는 requires_grad=False이므로 x^2에 1대입한 1이 나옴
# w=w.detach() # 당연히 에러 발생

w.backward()
print(x.grad)
```

```python
# 많이 쓰이는 torch.no_grad
x=torch.tensor([1.], requires_grad=True)
# chain rule을 위해 계속 grad_fn을 update 하니까 grad_fn 잠시 안 계산하고 싶을 때 torch.no_grad
# 모델 테스트 시에는 불필요하게 메모리 쓸 필요 없기 때문!

"""
requires_grad가 False로 잠시 바뀐거냐? : X(여전히 True)
why? torch.no_grad()에 감싸져 있기 때문
"""
with torch.no_grad():
    y = x**2
    print(x.requires_grad)
    print(y)
print(x.requires_grad)
# y.backward() # error!

print()

y=x**2
print(y)
y.backward()
print(x.grad)

print()

x=torch.tensor([1.], requires_grad=True)
x.requires_grad=False
y = x**2
print(x.requires_grad)
print(y)
# y.backward() # error!
```

```python
from torch import nn

x=torch.tensor([1.])
print(x.shape)
model=nn.Linear(1,1) # 입력 node 한 개, 출력 node 한 개인 layer 만듦
print(model)

print(model.weight.shape) # 만들면서 initialize(초기) 함
print(model.bias)

y=model(x)
print(y)

y2 = x @ model.weight + model.bias # 식으로 나타내 본다면
print(y2)

make_dot(y)
```

```python
model = nn.Linear(2,3)
print(model.weight)
print(model.bias)
print()
x = torch.randn(2)
print(x)
print(model(x))

"""
nn.Linear는 데이터의 shape의 마지막 차원이 '채'로(1D data) 들어오길 기대하는 녀석이다
입력 노드 하나가 곧 하나의 채널(피처) 값을 의미
'채널'은 'TV 채널'의 '채널' 같이 특정 유형의 정보를 전달하는 통로. 즉, 피처(특징)와 의미적으로 거의 비슷함
"""

y2 = x@model.weight.T + model.bias
print(y2)

make_dot(model(x))
```

```python
class MyModel(nn.Module):
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5)
        self.fc2 = nn.Linear(5,10)
        self.fc3 = nn.Linear(10,3)
        self.act = nn.Sigmoid()


    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.act(x)
        x = self.fc3(x)
        x = self.act(x)
        return x

model = MyModel()
x = torch.randn(5,2)
y = model(x) # model.forward(x) (nn.Module의 __call__에서 forward를 통과시킴) (__call_은 인스턴스()로 호출하는 메서드)
print(y)

label = ['하','중','상']

pred = torch.argmax(y, dim=1)
print(pred)

label_pred = [label[i] for i in pred]
print(label_pred)
# make_dot(y)


# import torch
# import torch.nn as nn
# import torch.nn.functional as F

# class MyModel(nn.Module):
#     def __init__(self):
#         super().__init__()
#         self.fc1 = nn.Linear(2, 5)
#         self.fc2 = nn.Linear(5, 10)
#         self.fc3 = nn.Linear(10, 3)
#         self.act = nn.Sigmoid()  # 중간층 활성화로 Sigmoid 사용 가능

#     def forward(self, x):
#         x = self.fc1(x)
#         x = self.act(x)
#         x = self.fc2(x)
#         x = self.act(x)
#         x = self.fc3(x)
#         x = F.softmax(x, dim=1)  # 마지막에 softmax 적용하여 확률 출력
#         return x

# model = MyModel()
# x = torch.randn(5, 2)
# y = model(x)
# print("Softmax 확률 출력:\n", y)

# pred = torch.argmax(y, dim=1)
# print("예측 클래스 인덱스:", pred)

# label = ['하', '중', '상']
# label_pred = [label[i] for i in pred]
# print("예측 클래스 이름:", label_pred)
```
