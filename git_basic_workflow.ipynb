## Git 파일 삭제

```python
# 옵션 1: 파일 완전히 삭제하려면 (로컬에서도 삭제)
# !git rm -f "filename"
```

```python
# # 옵션 2: GitHub에서만 삭제하고, 로컬 파일은 남기고 싶다면
!git rm --cached "filename"
```

```python
# 3. 커밋 메시지 작성
!git commit -m "delete filename"
```

```python
# 4. 원격 저장소에서 최신 커밋을 가져와서 리베이스
!git pull origin main --rebase
```

```python
# 5. 변경사항을 원격 저장소로 푸시
!git push origin main
```
